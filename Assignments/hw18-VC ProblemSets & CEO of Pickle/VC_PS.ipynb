```python
# Recalculating all key figures with a 10% stock option pool included at liquidity event.

# Given values:
terminal_value = 91000000  # at liquidity event
investment_A = 1500000
investment_B = 7000000
investment_C = 8500000
shares_founders = 500000  # initial founders shares
shares_SAFE = 55556  # from previous calculation
old_shares_before_A = shares_founders + shares_SAFE

# Adjusting for stock option pool (10% of company at exit), new final percentage = 90%
adjusted_terminal_value = terminal_value

# Calculate required ownership percentages at exit (same as previously calculated)
VI_A = investment_A * (1 + 0.50)**5
VI_B = investment_B * (1 + 0.40)**3
VI_C = investment_C * (1 + 0.30)**2

# Ownership percentages at exit
ownership_A_exit = VI_A / adjusted_terminal_value
ownership_B_exit = VI_B / adjusted_terminal_value
ownership_C_exit = VI_C / adjusted_terminal_value

# Adjust ownership for future dilution and option pool (10% reserved at end)
retention_A = 1 - (ownership_B_exit + ownership_C_exit + 0.10)
retention_B = 1 - (ownership_C_exit + 0.10)
retention_C = 1 - 0.10

ownership_A_initial = ownership_A_exit / retention_A
ownership_B_initial = ownership_B_exit / retention_B
ownership_C_initial = ownership_C_exit / retention_C

# New shares for Series A
new_shares_A = (ownership_A_initial / (1 - ownership_A_initial)) * old_shares_before_A
total_shares_after_A = old_shares_before_A + new_shares_A
price_per_share_A = investment_A / new_shares_A
post_money_A = total_shares_after_A * price_per_share_A
pre_money_A = post_money_A - investment_A

# Series B calculations
new_shares_B = (ownership_B_initial / (1 - ownership_B_initial)) * total_shares_after_A
total_shares_after_B = total_shares_after_A + new_shares_B
price_per_share_B = investment_B / new_shares_B
post_money_B = total_shares_after_B * price_per_share_B
pre_money_B = post_money_B - investment_B

# Series C calculations
new_shares_C = (ownership_C_initial / (1 - ownership_C_initial)) * total_shares_after_B
total_shares_after_C = total_shares_after_B + new_shares_C
price_per_share_C = investment_C / new_shares_C
post_money_C = total_shares_after_C * price_per_share_C
pre_money_C = post_money_C - investment_C

# Total shares including 10% stock option pool at exit
final_total_shares = total_shares_after_C / (1 - 0.10)

# Terminal price per share at liquidity event
terminal_share_price = terminal_value / final_total_shares

# Valuations for Ken, other founders, and Cathy at liquidity event
ken_shares_value = 350000 * terminal_share_price
cofounder_shares_value = 75000 * terminal_share_price
cathy_shares_value = shares_SAFE * terminal_share_price

# Calculate compounded annual return
years_total = 6
ken_roi = (ken_shares_value / 30000)**(1/years_total) - 1
cofounder_roi = (cofounder_shares_value / 10000)**(1/years_total) - 1
cathy_roi = (cathy_shares_value / 400000)**(1/years_total) - 1

results = {
    "Series_A": {
        "ownership_initial": ownership_A_initial,
        "new_shares": new_shares_A,
        "price_per_share": price_per_share_A,
        "pre_money": pre_money_A,
        "post_money": post_money_A
    },
    "Series_B": {
        "ownership_initial": ownership_B_initial,
        "new_shares": new_shares_B,
        "price_per_share": price_per_share_B,
        "pre_money": pre_money_B,
        "post_money": post_money_B
    },
    "Series_C": {
        "ownership_initial": ownership_C_initial,
        "new_shares": new_shares_C,
        "price_per_share": price_per_share_C,
        "pre_money": pre_money_C,
        "post_money": post_money_C
    },
    "Exit": {
        "final_total_shares": final_total_shares,
        "terminal_share_price": terminal_share_price,
        "ken_shares_value": ken_shares_value,
        "cofounder_shares_value": cofounder_shares_value,
        "cathy_shares_value": cathy_shares_value,
        "ken_roi": ken_roi,
        "cofounder_roi": cofounder_roi,
        "cathy_roi": cathy_roi
    }
}

results
```

```python
# Scenario #6: Participating Preferred Stock for Series B & C (original amounts, no delay, option pool included from Q4)

# Original values from Q4 scenario
terminal_value = 91000000
years_to_exit = 5  # Series B (2027-2030 = 3 yrs), Series C (2028-2030 = 2 yrs), Series A (2025-2030 = 5 yrs)

# Original Series B and C investment amounts (from Q4)
investment_B_original = 7000000
investment_C_original = 8500000

# Original shares acquired in Q4 scenario
shares_A = results["Series_A"]["new_shares"]
shares_B = results["Series_B"]["new_shares"]
shares_C = results["Series_C"]["new_shares"]

# Initial payback from participating preferred stock (Series B & C)
priority_payment = investment_B_original + investment_C_original
remaining_value = terminal_value - priority_payment

# Distributions to each investor class after priority payments
total_shares_at_exit = results["Exit"]["final_total_shares"]
terminal_share_price_adj = remaining_value / total_shares_at_exit

# Valuation of each series after priority payments
value_series_B_exit_adj = priority_payment * (investment_B_original / (investment_B_original + investment_C_original)) + (shares_B * terminal_share_price_adj)
value_series_C_exit_adj = priority_payment * (investment_C_original / (investment_B_original + investment_C_original)) + (shares_C * terminal_share_price_adj)
value_series_A_exit_adj = shares_A * terminal_share_price_adj

# ROI calculations adjusted for Series B & C participating preferred
roi_series_B_pp = (value_series_B_exit_adj / investment_B_original)**(1/3) - 1
roi_series_C_pp = (value_series_C_exit_adj / investment_C_original)**(1/2) - 1
roi_series_A_adj = (value_series_A_exit_adj / investment_A)**(1/5) - 1

# Founder and Cathy share distributions at exit
founders_shares_total = shares_founders
cathy_shares_total = shares_SAFE

founders_distribution = founders_shares_total * terminal_share_price_adj
cathy_distribution = cathy_shares_total * terminal_share_price_adj

results_pp_stock = {
    "Adjusted_ROIs": {
        "Series_A_ROI": roi_series_A_adj,
        "Series_B_ROI": roi_series_B_pp,
        "Series_C_ROI": roi_series_C_pp
    },
    "Distributions": {
        "Founders": founders_distribution,
        "Cathy": cathy_distribution
    }
}

results_pp_stock
```
