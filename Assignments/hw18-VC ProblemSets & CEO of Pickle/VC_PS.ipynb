## Problem Overview

XL Microdevices is raising funding through several financing rounds (Series A, B, C). 


```python
import numpy as np
import pandas as pd
```

```python
# Initial Seed Investments
founders_investment = {'Ken Cook': 30000, 'Founder 2': 10000, 'Founder 3': 10000}
founders_shares = {'Ken Cook': 350000, 'Founder 2': 75000, 'Founder 3': 75000}

# SAFE  
safe_investment = 400000
safe_discount = 0.20
safe_valuation_cap = 4000000
safe_year = 2024

investments = {
    'Series A': {'amount': 1500000, 'year': 2025, 'target_return': 0.50},
    'Series B': {'amount': 7000000, 'year': 2027, 'target_return': 0.40},
    'Series C': {'amount': 8500000, 'year': 2028, 'target_return': 0.30},
}

exit_year = 2030
expected_revenue = 6500000
price_revenue_ratio = 14
exit_valuation = expected_revenue * price_revenue_ratio
```

```python
years_to_exit_A = exit_year - investments['Series A']['year']

required_exit_value_A = investments['Series A']['amount'] * (1 + investments['Series A']['target_return']) ** years_to_exit_A

# Series A ownership percentage
series_A_ownership = required_exit_value_A / exit_valuation

post_money_A = investments['Series A']['amount'] / series_A_ownership
pre_money_A = post_money_A - investments['Series A']['amount']

existing_shares = sum(founders_shares.values())
price_per_share_A = pre_money_A / existing_shares
new_shares_A = investments['Series A']['amount'] / price_per_share_A

print({
    'Series A Ownership %': round(series_A_ownership, 3),
    'New Shares Issued': round(new_shares_A, 3),
    'Price per Share': round(price_per_share_A, 3),
    'Pre-Money Valuation': round(pre_money_A, 3),
    'Post-Money Valuation': round(post_money_A, 3)
})
```

## Series B and C

```python

years_to_exit_B = exit_year - investments['Series B']['year']
required_exit_value_B = investments['Series B']['amount'] * (1 + investments['Series B']['target_return'])**years_to_exit_B
series_B_ownership = required_exit_value_B / exit_valuation

years_to_exit_C = exit_year - investments['Series C']['year']
required_exit_value_C = investments['Series C']['amount'] * (1 + investments['Series C']['target_return'])**years_to_exit_C
series_C_ownership = required_exit_value_C / exit_valuation
```

## Q4: Ken expand pool by 10%

```python

retention_A = 1 - (series_B_ownership + series_C_ownership + 0.10)
retention_B = 1 - (series_C_ownership + 0.10)
retention_C = 1 - 0.10

ownership_A_initial = series_A_ownership / retention_A
ownership_B_initial = series_B_ownership / retention_B
ownership_C_initial = series_C_ownership / retention_C

# Step 5: Calculate new shares for each round
shares_founders = sum(founders_shares.values())
shares_SAFE = 55556  # This is from previous calculations
old_shares_before_A = shares_founders + shares_SAFE

# Series A new shares
new_shares_A = (ownership_A_initial / (1 - ownership_A_initial)) * old_shares_before_A
total_shares_after_A = old_shares_before_A + new_shares_A

# Series B new shares
new_shares_B = (ownership_B_initial / (1 - ownership_B_initial)) * total_shares_after_A
total_shares_after_B = total_shares_after_A + new_shares_B

# Series C new shares
new_shares_C = (ownership_C_initial / (1 - ownership_C_initial)) * total_shares_after_B
total_shares_after_C = total_shares_after_B + new_shares_C

# Include the 10% stock option pool
final_total_shares = total_shares_after_C / (1 - 0.10)
terminal_share_price = exit_valuation / final_total_shares

ken_shares_value = founders_shares['Ken Cook'] * terminal_share_price
cofounder_shares_value = founders_shares['Founder 2'] * terminal_share_price
cathy_shares_value = shares_SAFE * terminal_share_price

# ROI
years_total = 6  # From 2024 to 2030
ken_roi = (ken_shares_value / founders_investment['Ken Cook'])**(1/years_total) - 1
cofounder_roi = (cofounder_shares_value / founders_investment['Founder 2'])**(1/years_total) - 1
cathy_roi = (cathy_shares_value / safe_investment)**(1/years_total) - 1

print("\n=== Investment Round Details ===")
print(f"Series A:")
print(f"  - Ownership: {ownership_A_initial:.2%}")
print(f"  - New Shares: {new_shares_A:,.0f}")
print(f"  - Pre-money Valuation: ${pre_money_A:,.2f}")

print(f"\nSeries B:")
print(f"  - Ownership: {ownership_B_initial:.2%}")
print(f"  - New Shares: {new_shares_B:,.0f}")

print(f"\nSeries C:")
print(f"  - Ownership: {ownership_C_initial:.2%}")
print(f"  - New Shares: {new_shares_C:,.0f}")

print("\n=== Exit Details ===")
print(f"Final Total Shares: {final_total_shares:,.0f}")
print(f"Terminal Share Price: ${terminal_share_price:.2f}")

print("\n=== Founder Returns ===")
print(f"Ken's Shares Value: ${ken_shares_value:,.2f}")
print(f"Ken's ROI: {ken_roi:.2%}")
print(f"\nCo-founder's Shares Value: ${cofounder_shares_value:,.2f}")
print(f"Co-founder's ROI: {cofounder_roi:.2%}")
print(f"\nCathy's Shares Value: ${cathy_shares_value:,.2f}")
print(f"Cathy's ROI: {cathy_roi:.2%}")
```

## Q5: Deley event for two years

Question #5. Immediately before the Series B round, it becomes apparent that the liquidity event will be delayed two years until March 2032 and that an additional $2,000,000 (i.e., a total of $10,500,000) will be needed in the scheduled Series C round. Although the liquidity event will be delayed two years, the Series B and Series C rounds of investments remain scheduled for March 2027 and March 2028, respectively. Despite the delay, the estimated terminal value of XL remains unchanged. At the time this delay becomes apparent, TVC’s Series A investment is already a done-deal and cannot be renegotiated. The future stock option pool, as described above, is included in XL’s plans. How does this change your answers to question #4? Given this delay scenario, what compound annual rates of return are realized on the Series A, B and C investments?


```python
# New conditions
new_exit_year = 2032  # 2 years later
increased_C_amount = 10500000  # Increased by $2M
terminal_value = 91000000  # Same as before

years_to_exit_A_new = new_exit_year - investments['Series A']['year']  
years_to_exit_B_new = new_exit_year - investments['Series B']['year']  
years_to_exit_C_new = new_exit_year - investments['Series C']['year'] 

VI_A_new = investments['Series A']['amount'] * (1 + investments['Series A']['target_return'])**years_to_exit_A_new
VI_B_new = investments['Series B']['amount'] * (1 + investments['Series B']['target_return'])**years_to_exit_B_new
VI_C_new = increased_C_amount * (1 + investments['Series C']['target_return'])**years_to_exit_C_new

ownership_A_exit_new = VI_A_new / terminal_value
ownership_B_exit_new = VI_B_new / terminal_value
ownership_C_exit_new = VI_C_new / terminal_value

#  10% option pool at exit
retention_B_new = 1 - (ownership_C_exit_new + 0.10)
retention_C_new = 1 - 0.10

# Series A is locked in at its original initial ownership
# Calculate new initial ownership for B and C
ownership_B_initial_new = ownership_B_exit_new / retention_B_new
ownership_C_initial_new = ownership_C_exit_new / retention_C_new

new_shares_B_new = (ownership_B_initial_new / (1 - ownership_B_initial_new)) * total_shares_after_A
total_shares_after_B_new = total_shares_after_A + new_shares_B_new
price_per_share_B_new = investments['Series B']['amount'] / new_shares_B_new

new_shares_C_new = (ownership_C_initial_new / (1 - ownership_C_initial_new)) * total_shares_after_B_new
total_shares_after_C_new = total_shares_after_B_new + new_shares_C_new
price_per_share_C_new = increased_C_amount / new_shares_C_new

final_total_shares_new = total_shares_after_C_new / (1 - 0.10)
terminal_share_price_new = terminal_value / final_total_shares_new

value_A_exit = new_shares_A * terminal_share_price_new
value_B_exit = new_shares_B_new * terminal_share_price_new
value_C_exit = new_shares_C_new * terminal_share_price_new

roi_A_new = (value_A_exit / investments['Series A']['amount'])**(1/years_to_exit_A_new) - 1
roi_B_new = (value_B_exit / investments['Series B']['amount'])**(1/years_to_exit_B_new) - 1
roi_C_new = (value_C_exit / increased_C_amount)**(1/years_to_exit_C_new) - 1

# Founder returns
ken_shares_value = founders_shares['Ken Cook'] * terminal_share_price_new
cofounder_shares_value = founders_shares['Founder 2'] * terminal_share_price_new
cathy_shares_value = shares_SAFE * terminal_share_price_new

years_total_new = new_exit_year - 2024
ken_roi = (ken_shares_value / founders_investment['Ken Cook'])**(1/years_total_new) - 1
cofounder_roi = (cofounder_shares_value / founders_investment['Founder 2'])**(1/years_total_new) - 1
cathy_roi = (cathy_shares_value / safe_investment)**(1/years_total_new) - 1

# Print the results
print("\n=== QUESTION 5: DELAYED EXIT SCENARIO ===")
print(f"Series C investment increased to ${increased_C_amount:,.0f}")

print("\n=== Investment Round Details ===")
print(f"Series A (locked in):")
print(f"  - Ownership: {ownership_A_initial:.2%}")
print(f"  - New Shares: {new_shares_A:,.0f}")

print(f"\nSeries B (adjusted):")
print(f"  - Ownership: {ownership_B_initial_new:.2%}")
print(f"  - New Shares: {new_shares_B_new:,.0f}")
print(f"  - Price per Share: ${price_per_share_B_new:.2f}")

print(f"\nSeries C (increased amount):")
print(f"  - Ownership: {ownership_C_initial_new:.2%}")
print(f"  - New Shares: {new_shares_C_new:,.0f}")
print(f"  - Price per Share: ${price_per_share_C_new:.2f}")

print("\n=== Compound Annual ROIs ===")
print(f"Series A ROI: {roi_A_new:.2%}")
print(f"Series B ROI: {roi_B_new:.2%}")
print(f"Series C ROI: {roi_C_new:.2%}")

print("\n=== Founder Returns ===")
print(f"Ken's Shares Value: ${ken_shares_value:,.2f}")
print(f"Ken's Annual ROI: {ken_roi:.2%}")
print(f"\nCo-founder's Shares Value: ${cofounder_shares_value:,.2f}")
print(f"Co-founder's Annual ROI: {cofounder_roi:.2%}")
print(f"\nCathy's Shares Value: ${cathy_shares_value:,.2f}")
print(f"Cathy's Annual ROI: {cathy_roi:.2%}")
```

# Q6: Participating Preferred Stock for Series B & C



```python
priority_payment = investments['Series B']['amount'] + investments['Series C']['amount']
remaining_value = exit_valuation - priority_payment

adjusted_share_price = remaining_value / final_total_shares

value_series_A = new_shares_A * adjusted_share_price

b_portion = investments['Series B']['amount'] / (investments['Series B']['amount'] + investments['Series C']['amount'])
value_series_B = (priority_payment * b_portion) + (new_shares_B * adjusted_share_price)

c_portion = investments['Series C']['amount'] / (investments['Series B']['amount'] + investments['Series C']['amount'])
value_series_C = (priority_payment * c_portion) + (new_shares_C * adjusted_share_price)

roi_series_A_pp = (value_series_A / investments['Series A']['amount'])**(1/years_to_exit_A) - 1
roi_series_B_pp = (value_series_B / investments['Series B']['amount'])**(1/years_to_exit_B) - 1
roi_series_C_pp = (value_series_C / investments['Series C']['amount'])**(1/years_to_exit_C) - 1

ken_distribution = founders_shares['Ken Cook'] * adjusted_share_price
founder2_distribution = founders_shares['Founder 2'] * adjusted_share_price
founder3_distribution = founders_shares['Founder 3'] * adjusted_share_price
cathy_distribution = shares_SAFE * adjusted_share_price

total_founder_distribution = ken_distribution + founder2_distribution + founder3_distribution

print("\n=== QUESTION 6: PARTICIPATING PREFERRED STOCK ===")
print("Series B and C will first get their original investment back ($15.5M)")
print(f"Then all shareholders participate in the remaining ${remaining_value:,.2f}")

print("\n=== TVC Returns with Participating Preferred ===")
print("Series A (standard preferred):")
print(f"  - Exit Value: ${value_series_A:,.2f}")
print(f"  - Annual ROI: {roi_series_A_pp:.2%}")

print("\nSeries B (participating preferred):")
print(f"  - Priority Payment: ${priority_payment * b_portion:,.2f}")
print(f"  - Share Value: ${new_shares_B * adjusted_share_price:,.2f}")
print(f"  - Total Value: ${value_series_B:,.2f}")
print(f"  - Annual ROI: {roi_series_B_pp:.2%}")

print("\nSeries C (participating preferred):")
print(f"  - Priority Payment: ${priority_payment * c_portion:,.2f}")
print(f"  - Share Value: ${new_shares_C * adjusted_share_price:,.2f}")
print(f"  - Total Value: ${value_series_C:,.2f}")
print(f"  - Annual ROI: {roi_series_C_pp:.2%}")

print("\n=== Founder & SAFE Distributions ===")
print(f"Ken's Distribution: ${ken_distribution:,.2f}")
print(f"Founder 2's Distribution: ${founder2_distribution:,.2f}")
print(f"Founder 3's Distribution: ${founder3_distribution:,.2f}")
print(f"Total Founders: ${total_founder_distribution:,.2f}")
print(f"Cathy's Distribution: ${cathy_distribution:,.2f}")

standard_share_price = exit_valuation / final_total_shares
standard_founders = shares_founders * standard_share_price
standard_cathy = shares_SAFE * standard_share_price

print("\n=== Comparison with Standard Preferred ===")
print(f"Founders with Standard Preferred: ${standard_founders:,.2f}")
print(f"Founders with Participating Preferred: ${total_founder_distribution:,.2f}")
print(f"Difference: ${standard_founders - total_founder_distribution:,.2f} ({(1 - total_founder_distribution/standard_founders):.2%} less)")

print(f"\nCathy with Standard Preferred: ${standard_cathy:,.2f}")
print(f"Cathy with Participating Preferred: ${cathy_distribution:,.2f}")
print(f"Difference: ${standard_cathy - cathy_distribution:,.2f} ({(1 - cathy_distribution/standard_cathy):.2%} less)")
```
